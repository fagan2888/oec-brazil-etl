```python
import pandas as pd
import numpy as np

all_p = np.empty([1,0])

for f in ["EXP","IMP"]:
    for y in [str(x) for x in range(1997,2020)]:
        filename = "{}_{}_MUN.csv".format(f,y)
        df = pd.read_csv("./data_temp/"+filename, sep=";")["CO_PAIS"]
        c = np.array(df.values)
        print("{} {} ".format(f,y), c)
        all_p = np.append(all_p, c)
        
p = np.unique(all_p)
p
```

```python
g1 = pd.read_csv("./resources/PAIS.csv", sep=";", encoding="latin-1")
g1_dict = {k:v for (k,v) in zip(g1["CO_PAIS"],g1["CO_PAIS_ISOA3"])}

g2 = pd.read_csv("./resources/shared_countries.csv")
g2_dict = {k:v for (k,v) in zip(g2["iso3"],g2["id_num"])}

u = [int(x) for x in p]
u = pd.Series(u).map(g1_dict).str.lower()
u = pd.DataFrame({"iso3":u})
u = u.merge(g2[["iso3","id_num"]], on="iso3", how="left")
u[u["id_num"].isnull()]
```
